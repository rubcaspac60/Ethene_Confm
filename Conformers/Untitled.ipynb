```python
import sys
from pyscf import gto, scf
from pyscf import lo
import numpy



def molecule_dyn(xyz, basis, spin, charge, verbose): 
    from pyscf import gto
    mol = gto.Mole()
    mol.fromfile(str(xyz)) 
    mol.basis = basis   # 'sto-3g' , 'sto-6g', 'cc-aug-TZVP' and 'def2-TZVP'
    mol.symmetry = True   # T / F
    mol.spin = spin    # 0 / 1
    mol.charge = charge  
    mol.unit = 'A'   # for bohr use 'B'
    mol.verbose = verbose   # No output
    #mol.output = '/Users/rcp/Documents/QiskitHackatonEurope/Ethene_Results_Runtime.txt'
    mol.build()
    return mol

def optimize(mol):
    from pyscf import gto
    from pyscf.geomopt.geometric_solver import optimize
    conv_params = { # These are the default settings
    'convergence_energy': 1e-8,  # Eh
    'convergence_grms': 2e-4,    # Eh/Bohr
    'convergence_gmax': 2.5e-4,  # Eh/Bohr
    'convergence_drms': 0.5e-3,  # Angstrom
    'convergence_dmax': 0.5e-3,  # Angstrom
}
    opt= optimize(mol).run()
    return opt.tofile("opt_C2H4_Dimer_12.xyz")
```
